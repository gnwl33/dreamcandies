```python
import csv
import time
import random
            
def write_customer_sample():
    with open('customer_sample.csv') as df1:
        data1 = csv.reader(df1)
        next(data1)
        with open('customer.csv', mode='a') as df2:
            data2 = csv.writer(df2, quoting=csv.QUOTE_ALL)
            for row in data1:
                new_row = row + ['John', 'Smith']
                data2.writerow(new_row)
                
def add_zeros(num):
    res = ''
    while num > 0:
        res += '0'
        num -= 1
    return res

def write_customer_data(name, fields, end):
    with open('customer_sample.csv') as df1:
        data1 = csv.reader(df1)
        next(data1)
        samples = set()
        for row in df1:
            samples.add(int(row[5:-2]))
        with open('customer.csv'.format(name), mode='a') as df2:
            data2 = csv.writer(df2, quoting=csv.QUOTE_ALL)
            for i in range(500000):
                if i not in samples:
                    digits = len(str(i))
                    key = 'CUST' + add_zeros(10 - digits) + str(i)
                    data2.writerow([key, 'Jane', 'Doe'])
                    i += 1

def write_invoice_data():
    with open('customer.csv') as df1:
        data1 = csv.reader(df1)
        next(data1)
        with open('invoice.csv', mode='a') as df2:
            data2 = csv.writer(df2, quoting=csv.QUOTE_ALL)
            count = 0
            for row in data1:
                if row[1] == 'Jane':
                    for i in range(2):
                        digits = len(str(count))
                        key = 'IN' + add_zeros(7 - digits) + str(count)
                        data2.writerow([row[0], key, "105.50", "01-Jan-2016"])
                        count += 1
                else:
                    for i in range(2):
                        digits = len(str(count))
                        key = 'IN' + add_zeros(7 - digits) + str(count)
                        data2.writerow([row[0], key, "114.14", "01-Feb-2017"])
                        count += 1
            
def write_invoice_item_data():
    with open('invoice.csv') as df1:
        data1 = csv.reader(df1)
        next(data1)
        with open('invoice_item.csv', mode='a') as df2:
            data2 = csv.writer(df2, quoting=csv.QUOTE_ALL)
            for row in data1:
                if row[2] == '105.50':
                    for i in range(5):
                        data2.writerow([row[1], 'MEIJI', "75.60", "100"])
                else:
                    for i in range(5):
                        data2.writerow([row[1], 'POCKY', "10.40", "250"])

start_time = time.time()
# write_customer_sample()
# write_customer_data()
# write_invoice_data()
write_invoice_item_data()
print("--- %s seconds ---" % (time.time() - start_time))
```
